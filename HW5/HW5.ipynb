用 Keras 建構自己的模型，進行鐵達尼號生存分析

```python
import pandas as pd
import numpy as np
import keras
from keras.layers import Dense , Dropout
from keras.models import Sequential
```

## 讀取資料

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
submit = pd.read_csv('gender_submission.csv')
```

```python
train.head()
```

```python
test.head()
```

## 選取features & 將train資料分割為訓練集與測試集

```python
selected_features = ['Pclass','Sex','Age','SibSp','Parch','Fare']
```

```python
x = train[selected_features]   
y = train['Survived']          

x_predict = test[selected_features]      #最後用來預測的測試集
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5278)
```

## 處理缺失值

```python
x_train.info()
```

```python
x_test.info()
```

```python
x_predict.info()
```

```python
x_train['Age'].fillna(x_train['Age'].mean(), inplace=True)
x_test['Age'].fillna(x_test['Age'].mean(), inplace=True)

x_predict['Age'].fillna(x_predict['Age'].mean(), inplace=True)
x_predict['Fare'].fillna(x_test['Fare'].mean(), inplace=True)
```

## 將字串類型的類別資料做轉換(將Sex轉為0,1)

```python
x_train['Sex'] = x_train['Sex'].map({'male':0,'female':1})
x_test['Sex'] = x_test['Sex'].map({'male':0,'female':1})
x_predict['Sex'] = x_predict['Sex'].map({'male':0,'female':1})
```

## 利用keras建構模型

```python
model = keras.models.Sequential()
```

```python
model.add(keras.layers.Dense(200,kernel_initializer='uniform',input_dim=6 , activation=('relu')))    #隱藏層1
```

```python
model.add(keras.layers.Dense(300,kernel_initializer='uniform',activation=('relu')))                  #隱藏層2
```

```python
model.add(keras.layers.Dense(1,kernel_initializer='uniform',activation=('sigmoid')))                 #輸出層
```

## 訓練模型

```python
model.compile(loss='mse', optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train, y_train,epochs=800)
```

## 訓練集準確率

```python
acc_train = model.evaluate(x_train, y_train)
```

```python
acc_train[1]
```

## 測試集準確率

```python
acc_test = model.evaluate(x_test, y_test)
```

```python
acc_test[1]
```

## 利用訓練好的模型預測生存

```python
y_predict = model.predict(x_predict)
```

```python
y_predict1 = np.where(y_predict > 0.9 , 1 ,0)
y_predict2 = y_predict1.flatten()     #將y_predict轉成1-dimension
y_predict2
```

```python
submission = pd.DataFrame({'PassengerId':test['PassengerId'], 'Survived':y_predict2})
```

```python
submission.to_csv('predict_submission.csv', index=False)
```
