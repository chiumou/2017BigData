# group_by計算investment growth rate

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import itertools as it
```

```python
df = pd.read_csv("Stock_data.csv", encoding="Big5", header=0)
```

```python
a = df.sort_values(['code','Year'])
a
```

# <法一> 多一行data：去年投資額(invest_lag)

```python
def lag(data):
    data['invest_lag'] = data['invest'].shift(1)
    return data
```

```python
b = a.groupby(['code'])
c = b.apply(lag)
c['invest_growth'] = (c['invest']/c['invest_lag'])-1
c[c['code']==2330]
```

# <法二> 直接在函式中計算，不用多一行invest_lag

```python
def happy(data):
    lag = data['invest'].shift(1)
    data['invest_growth'] = (data['invest']/lag)-1
    return data
```

```python
b = a.groupby(['code'])
c = b.apply(happy)
c[c['code']==2330]
```

# <法三> pct_change()

```python
def cool(data):
    data['invest_growth'] = data['invest'].pct_change()
    return data
```

```python
b = a.groupby(['code'])
c = b.apply(cool)
c[c['code']==1101]
```
